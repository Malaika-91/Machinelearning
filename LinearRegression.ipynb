```python
import numpy as np

# Step 1: Input Data
X = np.array([1000, 1500, 2000, 2500, 3000])
y = np.array([150, 200, 250, 300, 350])

# Step 2: Add a column of 1s to X for the intercept term
X_b = np.c_[np.ones((len(X), 1)), X]   # [[1,1000], [1,1500], ...]

# Step 3: Apply Normal Equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Step 4: Print Results
beta_0 = theta_best[0]
beta_1 = theta_best[1]

print("Intercept (β₀):", beta_0)
print("Slope (β₁):", beta_1)

# Step 5: Predict for a new value
X_new = np.array([[1, 2600]])  # New house size
y_pred = X_new.dot(theta_best)
print("Predicted price for 2600 sq.ft house:", y_pred[0])
```

```python
# --- Simple Gradient Descent Example ---

# Data
x = [1, 2, 3]
y = [2, 4, 6]

# Initialize parameters (start with guesses)
b0 = 0   # intercept
b1 = 0   # slope

# Learning rate
alpha = 0.1

# Number of iterations
epochs = 50

# Gradient Descent
for i in range(epochs):
    y_pred = [b0 + b1 * xi for xi in x]  # predicted y
    error_b0 = sum([(y_pred[j] - y[j]) for j in range(len(x))]) / len(x)
    error_b1 = sum([(y_pred[j] - y[j]) * x[j] for j in range(len(x))]) / len(x)
    
    # Update parameters
    b0 = b0 - alpha * error_b0
    b1 = b1 - alpha * error_b1
    
    # Print progress every few steps
    if i % 10 == 0:
        print(f"Iteration {i}: b0={b0:.4f}, b1={b1:.4f}")

print("\nFinal values:")
print(f"b0 (intercept) = {b0:.4f}")
print(f"b1 (slope) = {b1:.4f}")
```

```python
import numpy as np

X = np.array([
    [1, 2, 9],
    [1, 3, 8],
    [1, 5, 6]
])

Y = np.array([
    [75],
    [82],
    [90]
])

XT = X.T

# Use pseudo-inverse instead of normal inverse
beta = np.linalg.pinv(XT.dot(X)).dot(XT.dot(Y))

print("Beta values (β₀, β₁, β₂):")
print(beta)

new_data = np.array([1, 4, 7])
predicted = new_data.dot(beta)
print("\nPredicted marks for 4 hrs study & 7 hrs sleep:")
print(predicted[0])
```
